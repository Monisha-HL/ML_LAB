```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# -----------------------------
# Task 1: Canada Per Capita Income Prediction
# -----------------------------
# Load dataset
canada_data = pd.read_csv(r"C:\Users\BMSCECSE-SH\Downloads\canada_per_capita_income.csv")

# Inspect data
print(canada_data.head())

# Features (Year) and Target (Income)
X_canada = canada_data[['year']]
y_canada = canada_data['per capita income (US$)']

# Build regression model
model_canada = LinearRegression()
model_canada.fit(X_canada, y_canada)

# Predict income for year 2020
income_2020 = model_canada.predict([[2020]])
print(f"Predicted Canada Per Capita Income in 2020: {income_2020[0]:.2f} US$")

# -----------------------------
# Task 2: Employee Salary Prediction
# -----------------------------
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
salary_data = pd.read_csv(r"C:\Users\BMSCECSE-SH\Downloads\salary.csv")

# Inspect columns
print(salary_data.head())
print(salary_data.dtypes)

# Ensure numeric types
salary_data['YearsExperience'] = pd.to_numeric(salary_data['YearsExperience'], errors='coerce')
salary_data['Salary'] = pd.to_numeric(salary_data['Salary'], errors='coerce')

# Drop rows with missing values
salary_data = salary_data.dropna()

# Features and target
X_salary = salary_data[['YearsExperience']]
y_salary = salary_data['Salary']

# Build regression model
model_salary = LinearRegression()
model_salary.fit(X_salary, y_salary)

# Predict salary for 12 years of experience using a DataFrame
salary_12yrs = model_salary.predict(pd.DataFrame({'YearsExperience':[12]}))
print(f"Predicted Salary for 12 years of experience: {salary_12yrs[0]:.2f}")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder

# -----------------------------
# Task 1: Hiring Data - Salary Prediction
# -----------------------------
# Load dataset
hiring_data = pd.read_csv(r"C:\Users\BMSCECSE-SH\Downloads\hiring.csv")

# Inspect column names
print("Columns in hiring.csv:", hiring_data.columns)

# Replace textual numbers like 'five' with actual numeric values if present
word_to_num = {
    'zero': 0, 'one': 1, 'two': 2, 'three': 3, 'four': 4,
    'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10
}

hiring_data['experience'] = hiring_data['experience'].replace(word_to_num)

# Convert relevant columns to numeric
hiring_data['experience'] = pd.to_numeric(hiring_data['experience'], errors='coerce')
hiring_data['test_score(out of 10)'] = pd.to_numeric(hiring_data['test_score(out of 10)'], errors='coerce')
hiring_data['interview_score(out of 10)'] = pd.to_numeric(hiring_data['interview_score(out of 10)'], errors='coerce')
hiring_data['salary($)'] = pd.to_numeric(hiring_data['salary($)'], errors='coerce')

# Fill missing values with column means (instead of dropping everything)
hiring_data = hiring_data.fillna(hiring_data.mean(numeric_only=True))

# Features and target
X_hiring = hiring_data[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y_hiring = hiring_data['salary($)']

# Build regression model
model_hiring = LinearRegression()
model_hiring.fit(X_hiring, y_hiring)

# Predictions
candidate1 = pd.DataFrame({'experience':[2], 'test_score(out of 10)':[9], 'interview_score(out of 10)':[6]})
candidate2 = pd.DataFrame({'experience':[12], 'test_score(out of 10)':[10], 'interview_score(out of 10)':[10]})

print("Predicted Salary (2yr exp, 9 test, 6 interview):", model_hiring.predict(candidate1)[0])
print("Predicted Salary (12yr exp, 10 test, 10 interview):", model_hiring.predict(candidate2)[0])

# -----------------------------
# Task 2: 1000 Companies Data - Profit Prediction
# -----------------------------
# Load dataset
companies_data = pd.read_csv(r"C:\Users\BMSCECSE-SH\Downloads\1000_companies.csv")

# Encode categorical 'State' column
le = LabelEncoder()
companies_data['State'] = le.fit_transform(companies_data['State'])

# Features and target
X_companies = companies_data[['R&D Spend', 'Administration', 'Marketing Spend', 'State']]
y_companies = companies_data['Profit']

# Build regression model
model_companies = LinearRegression()
model_companies.fit(X_companies, y_companies)

# Encode 'Florida' using same encoder
florida_code = le.transform(['Florida'])[0]

# Prediction
new_company = pd.DataFrame({
    'R&D Spend':[91694.48],
    'Administration':[515841.3],
    'Marketing Spend':[11931.24],
    'State':[florida_code]
})

print("Predicted Profit for given company:", model_companies.predict(new_company)[0])
```
