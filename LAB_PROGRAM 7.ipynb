```python
# Import required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Load dataset
data = pd.read_csv(r"C:\Users\Monisha H L\Downloads\income.csv")

# Separate features and target
# Target column is 'income_level'
X = data.drop("income_level", axis=1)
y = data["income_level"]

# Split dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Step 1: Build AdaBoost with n_estimators=10
ada_default = AdaBoostClassifier(n_estimators=10, algorithm="SAMME", random_state=42)
ada_default.fit(X_train, y_train)
y_pred_default = ada_default.predict(X_test)
default_score = accuracy_score(y_test, y_pred_default)

print(f"Default AdaBoost (10 trees) Accuracy: {default_score:.4f}")
print("Confusion Matrix (Default):")
print(confusion_matrix(y_test, y_pred_default))

# Step 2: Fine-tune number of trees
best_score = 0
best_n = 0
best_matrix = None

for n in range(10, 201, 10):  # test from 10 to 200 trees in steps of 10
    ada = AdaBoostClassifier(n_estimators=n, algorithm="SAMME", random_state=42)
    ada.fit(X_train, y_train)
    y_pred = ada.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    
    print(f"n_estimators={n}, Accuracy={score:.4f}")
    
    if score > best_score:
        best_score = score
        best_n = n
        best_matrix = confusion_matrix(y_test, y_pred)

print(f"\nBest Accuracy: {best_score:.4f} achieved with {best_n} trees")
print("Confusion Matrix (Best Model):")
print(best_matrix)
```
